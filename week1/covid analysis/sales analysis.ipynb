```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# ✅ Step 2: Load the dataset
df = pd.read_csv('../datasets/sales_data.csv')
```

```python
# ✅ Step 3: Preview data
print(df.head())
print(df.info())
```

```python
# ✅ Step 4: Add total_value column
df['total_value'] = df['quantity'] * df['price']
```

```python
# ✅ Step 5: Grouped analysis - Total revenue per product
product_sales = df.groupby('product')['total_value'].sum().reset_index()
print(product_sales)
```

```python
# ✅ Step 6: Visualize product-wise sales
plt.figure(figsize=(8,5))
sns.barplot(x='product', y='total_value', data=product_sales, palette='viridis')
plt.title('Total Sales by Product')
plt.xlabel('Product')
plt.ylabel('Revenue')
plt.tight_layout()
plt.show()


```

```python
# Customer-wise total spend
customer_sales = df.groupby('customer_id')['total_value'].sum().reset_index()
print(customer_sales)
# Sort top spenders
customer_sales = customer_sales.sort_values(by='total_value', ascending=False)

# Plot
plt.figure(figsize=(8,5))
sns.barplot(x='customer_id', y='total_value', data=customer_sales, palette='magma')
plt.title('Total Spend per Customer')
plt.xlabel('Customer ID')
plt.ylabel('Total Spent (₹)')
plt.tight_layout()
plt.show()
```
